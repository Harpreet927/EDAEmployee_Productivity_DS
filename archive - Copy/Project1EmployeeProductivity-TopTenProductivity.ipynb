```python
df = EmployeeTopTen.copy()
year=[]
for i in df['Joining Date'].str.split('-'):
    if int(i[1])<23:
        year.append('20'+str(i[1]))
    else:
        year.append('19'+str(i[1]))

df['Joining Date'] = pd.Series(year).astype('int')
df['Experience'] = df['Joining Date'].apply(lambda x: 2024 - x)
df.head()
```

```python
df_ = EmployeeTopTen.copy()
df_['Experience'] = df_['Joining Date'].str.split('-').apply(lambda x: x[1]).astype('int')
df_['Experience'] = df_['Experience'].apply(lambda x: 2024 - (x + 2000) if x <= 24 else 2024 - (x+1900))
df_.head()
```

```python
# Calculate the correlation of all features with 'SalePrice'
correlations = EmployeeTopTen.corr()['Salary'].sort_values(ascending=False)

# Display the top 10 features most correlated with 'SalePrice'
top_correlations = correlations[1:6]
print(top_correlations)
```

```python
# Select the top correlated features including SalePrice
selected_features = list(top_correlations.index) + ['Salary']

# Compute the correlations for the selected features
correlation_matrix = EmployeeTopTen[selected_features].corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Generate a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5, fmt=".2f", vmin=-1, vmax=1)

# Title
plt.title("Heatmap of Correlations among Top Features with Salary", fontsize=16)

# Show the heatmap
plt.show()
```
