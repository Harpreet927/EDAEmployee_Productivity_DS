```python
# Load the 'Employee Productivity and Satisfaction HR Data' dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

Employee = pd.read_csv('hr_dashboard_data.csv')
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.kdeplot(Employee['Salary'], color='blue', ax=ax, shade=True)

# add vertical line at mean
mean = Employee['Salary'].mean()
ax.axvline(mean, color='black', linestyle='--')

# add vertical line at min
min = Employee['Salary'].min()
ax.axvline(min, color='red', linestyle='--')

# add vertical line at max
max = Employee['Salary'].max()
ax.axvline(max, color='green', linestyle='--')



plt.show()

#Bimodularity
```

```python
top_ten_quan = Employee.Salary.quantile(0.9)
EmployeeTopTen = Employee[Employee['Salary'] > top_ten_quan]
EmployeeTopTen
```

```python
#number of employees in unique positions
total = EmployeeTopTen['Department'].value_counts().reset_index()
total.columns=['Department','Count']

total
```

```python
position_counts = EmployeeTopTen['Department'].value_counts()
position_counts.plot(kind='bar');
```

```python
sns.pairplot(EmployeeTopTen, vars=['Age', 'Salary', 'Projects Completed'], hue = 'Department', 
             palette='Set1', height=2.5, aspect=1.75)
 
# Display the plot
plt.show()
```

```python
sns.pairplot(Employee, vars=['Age', 'Salary', 'Projects Completed'], hue = 'Position', 
             palette='Set1', height=2.5, aspect=1.75)
 
# Display the plot
plt.show()
```

```python
sns.pairplot(Employee, vars=['Age', 'Salary', 'Projects Completed'], hue = 'Gender', 
             palette='Set1', height=2.5, aspect=1.75)
 
# Display the plot
plt.show()
```
