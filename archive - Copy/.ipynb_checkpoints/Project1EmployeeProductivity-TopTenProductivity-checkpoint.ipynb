```python
# Calculate the correlation of all features with 'SalePrice'
correlations = Employee.corr()['Salary'].sort_values(ascending=False)

# Display the top 10 features most correlated with 'SalePrice'
top_correlations = correlations[1:6]
print(top_correlations)
```

```python
# Select the top correlated features including SalePrice
selected_features = list(top_correlations.index) + ['Salary']

# Compute the correlations for the selected features
correlation_matrix = Employee[selected_features].corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Generate a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=.5, fmt=".2f", vmin=-1, vmax=1)

# Title
plt.title("Heatmap of Correlations among Top Features with Salary", fontsize=16)

# Show the heatmap
plt.show()
```

```python
df = Employee.copy()
year=[]
for i in df['Joining Date'].str.split('-'):
    if int(i[1])<23:
        year.append('20'+str(i[1]))
    else:
        year.append('19'+str(i[1]))

df['Joining Date'] = pd.Series(year).astype('int')
df['Experience'] = df['Joining Date'].apply(lambda x: 2024 - x)
df.head()
```

```python
df_ = Employee.copy()
df_['Experience'] = df_['Joining Date'].str.split('-').apply(lambda x: x[1]).astype('int')
df_['Experience'] = df_['Experience'].apply(lambda x: 2024 - (x + 2000) if x <= 24 else 2024 - (x+1900))
df_.head()
```

# Gender Analysis w/ Salary

```python
df_females = Employee.loc[Employee['Gender']=='Female']
df_females.head()
```

```python
female_by_dept = df_females.groupby('Department')[['Department', 'Salary']].mean().sort_values('Salary', ascending = False).reset_index()
female_by_dept
```

```python
color = ['lightblue', 'blue', 'purple', 'red', 'black', 'orange']
edgecolor = ['red', 'orange', 'darkblue', 'darkred', 'yellow', 'black']
fig, ax = plt.subplots()
ax.barh(female_by_dept['Department'], female_by_dept['Salary'],
       color=color,
       edgecolor=edgecolor,
       linewidth=3)


# Show graph
plt.show()
#Finance & IT have the highest AVG salary for Female.
```

```python
female_by_Pos = df_females.groupby('Position')[['Position', 'Salary']].mean().sort_values('Salary', ascending = False).reset_index()
female_by_Pos
```

```python
color = ['lightblue', 'blue', 'purple', 'red', 'black', 'orange']
edgecolor = ['red', 'orange', 'darkblue', 'darkred', 'yellow', 'black']
fig, ax = plt.subplots()
ax.barh(female_by_Pos['Position'], female_by_Pos['Salary'],
       color=color,
       edgecolor=edgecolor,
       linewidth=3)


# Show graph
plt.show()
#Managers & Team Leads have the highest AVG salary for Female.
```

```python
df_males = Employee.loc[Employee['Gender']=='Male']
df_males.head()
```

```python
maleby_dept = df_males.groupby('Department')[['Department', 'Salary']].mean().sort_values('Salary', ascending = False).reset_index()
maleby_dept
```

```python
color = ['lightblue', 'blue', 'purple', 'red', 'black', 'orange']
edgecolor = ['red', 'orange', 'darkblue', 'darkred', 'yellow', 'black']
fig, ax = plt.subplots()
ax.barh(maleby_dept['Department'], maleby_dept['Salary'],
       color=color,
       edgecolor=edgecolor,
       linewidth=3)


# Show graph
plt.show()
#HR & IT have the highest AVG salary for male. IT have about the same avg salary for both gender, however, 
#there is a difference in HR, females have the lowest avg salary whereas avg salary for HR in males is far greater. 
```

```python
maleby_Pos = df_males.groupby('Position')[['Position', 'Salary']].mean().sort_values('Salary', ascending = False).reset_index()
maleby_Pos
```

```python
color = ['lightblue', 'blue', 'purple', 'red', 'black', 'orange']
edgecolor = ['red', 'orange', 'darkblue', 'darkred', 'yellow', 'black']
fig, ax = plt.subplots()
ax.barh(maleby_Pos['Position'], maleby_Pos['Salary'],
       color=color,
       edgecolor=edgecolor,
       linewidth=3)


# Show graph
plt.show()
#Managers & Team Leads have the highest AVG salary for male and it similarly is the same for female.
```
