```python
# Load the 'Employee Productivity and Satisfaction HR Data' dataset
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

Employee = pd.read_csv('hr_dashboard_data.csv')
```

```python
Employee.describe()
```

What is the average salary?
What is the maximum salary?
What is the minimum salary?
What is the range of the top 10% of the salary? [top 10 percentile]
How many employees are within this top 10% group? {Ans: Should be 20}

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.kdeplot(Employee['Salary'], color='blue', ax=ax, shade=True)

# add vertical line at mean
mean = Employee['Salary'].mean()
ax.axvline(mean, color='black', linestyle='--')

# add vertical line at min
min = Employee['Salary'].min()
ax.axvline(min, color='red', linestyle='--')

# add vertical line at max
max = Employee['Salary'].max()
ax.axvline(max, color='green', linestyle='--')



plt.show()
```

```python
top_ten_quan = Employee.Salary.quantile(0.9)
EmployeeTopTen = Employee[Employee['Salary'] > top_ten_quan]
EmployeeTopTen
```

```python
EmployeeTopTen.describe()
```

```python
 # Plotting a basic histogram
plt.hist(EmployeeTopTen['Salary'], bins=10, color='skyblue', edgecolor='black')
 
# Adding labels and title
plt.xlabel('Salary')
plt.ylabel('Employee')
plt.title('Top Ten Percentile')
 
# Display the plot
plt.show()
```

```python
df_ = Employee[['Productivity (%)', 'Satisfaction Rate (%)']]
df_.boxplot(column = ['Productivity (%)', 'Satisfaction Rate (%)']); 
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(14, 7))
sns.histplot(Employee['Projects Completed'], bins=30, kde=True, color="blue")

# add vertical line at mean
mean_ = Employee['Projects Completed'].mean()
plt.axvline(mean_, color='black', linestyle='--')

# add vertical line at min
min_ = Employee['Projects Completed'].min()
plt.axvline(min_, color='red', linestyle='--')

# add vertical line at max
max_ = Employee['Projects Completed'].max()
plt.axvline(max_, color='green', linestyle='--')

# add vertical line at median
median_ = Employee['Projects Completed'].median()
plt.axvline(median_, color='orange', linestyle='--')
 
# Adding labels and title
plt.xlabel('Projects Completed')
plt.ylabel('Employee')
plt.title('Projects Completed')

 
# Display the plot
plt.show()
```

```python
plt.hist(Employee['Feedback Score'], bins=5) 

# add vertical line at mean
mean_ = Employee['Feedback Score'].mean()
plt.axvline(mean_, color='black', linestyle='--')


plt.xlabel('Feedback Score')
plt.ylabel('Frequency')
plt.title('Feedback Score', fontsize = 20)
```
